### Q1.

**Flask is a popular web application framework written in Python that allows developers to build web applications quickly and easily. Flask is known for its simplicity, flexibility, and lightweight nature, making it a popular choice for small to medium-sized web projects.**

**Advantage of Flask Framework:-**

1. Lightweight: Flask is a micro-framework that is very lightweight and doesn't come with all the bells.
2. Flexibility: Flask is highly flexible and can be customized to fit the needs of a wide range of web projects.
3. Easy to learn: Flask is easy to learn and use, especially for developers who are already familiar with Python.
4. Modular: Flask is highly modular and allows developers to add or remove components as needed.
5. Large community: Flask has a large and active community of developers, which means that there are plenty of resources and tools available to help developers build and maintain their applications.

**App routing in Flask refers to the process of mapping URLs to specific functions or views within a Flask application.**

**We use app routes in Flask to define the behavior of our web application. When a client requests a particular URL, Flask checks the routing configuration to see which view or function should handle the request. The view or function then processes the request, performs any necessary computations, and returns a response to the client.**
